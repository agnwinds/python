```python
from glob import glob
from astropy.table import Table,join,vstack
import numpy as np
import os
os.getcwd()
```

```python
def do_compare(doc_dir='/Users/long/Python/docs/sphinx/source/input/parameters/',template_dir='./parameters'):
    doc_files=files = glob('%s/**/*.rst' % doc_dir, recursive=True)
    print(len(doc_files))
    xname=[]
    for one in doc_files:
        xname.append(one.split('/')[-1])

    doc_tab=Table([xname,doc_files],names=['File','Path'])
    doc_tab['Doc']='Yes'
    
    template_files= glob ('%s/**/*.rst' % template_dir, recursive=True)
    print(len(template_files))
    yname=[]
    for one in template_files:
        yname.append(one.split('/')[-1])
    template_tab=Table([yname,template_files],names=['File','Path'])
    template_tab['Template']='Yes'
    names=np.append(np.array(doc_tab['File']),np.array(template_tab['File']))
    print(len(names))
    unique_names,counts=np.unique(names,return_counts=True)
    print(len(unique_names))
    for i in range(len(unique_names)):
        print(unique_names[i],counts[i])
    master_table=Table([unique_names], names=['File'])
    master_table=join(master_table,doc_tab,join_type='left',keys=['File'])
    master_table=join(master_table,template_tab,join_type='left',keys=['File'])
    return master_table
    
    
    
    
    
```

```python
master_table=do_compare()
```
